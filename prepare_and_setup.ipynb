## Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Read data

```python
input_path = '../../DATA/Commuting-Distance-by-Personal-Characteristics/generated_data/'
input_file = 'transformed_data.xlsx'
```

```python
df = pd.read_excel(input_path + input_file, usecols=[1, 2, 3], sheet_name='occupation')
df
```

The generated data contains the information about commuting distance in Australia in 2016. Column 'Occupation (OCCP)' includes different occupation types, while 'Average commuting distance (kilometres)' column shows the average commuting distance for these occupation types. The average commuting distance in Australia is 15.98km where "Machinery Operators and Drivers" occupation type has the highest average commuting distance (21.07km).   

## Visualize data

Here, we'd like to visualize the generated data. Below code demonstrates how to visualize the data by using Matplot library.

**Matplot** : Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python.

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')

x = list(df['Occupation (OCCP)'])[:-1]
AVG_distances = list(df['Average commuting distance \n(kilometres)'])[:-1]

x_pos = [i for i, _ in enumerate(x)]
plt.figure(figsize=(22,14))
plt.bar(x_pos, AVG_distances, color='#14AF43')
plt.xlabel("Occupation", fontsize=22)
plt.ylabel("Kilometres", fontsize=22)
plt.title("Average commuting distance (km) by Occupation", fontsize=28)
plt.ylim((0, 25))   # set the ylim to bottom, top
for index, value in enumerate(AVG_distances):
    plt.text(index - 0.18 , value - 1.8, str(round(value, 2)), fontsize=18)

plt.xticks(x_pos, x, fontsize=24, rotation=90)

plt.savefig('img/Commuting distance by occupation.pdf', bbox_inches='tight')
plt.show()
```

Across the different occupation types, "Machinery Operators and Drivers" occupation type requires the longest daily commute with an average commuting distance of 21.1km. On the other hand, "Sales Workers" occupation type has the lowest daily commute with an average commuting distance of 12.6km.

## Insight

**Sales workers have a relatively lower daily commuting distance (12.6km) compared to other occupations.**

```python
value = float(df['Average commuting distance \n(kilometres)'].loc[df['Occupation (OCCP)'] == 'Sales Workers'])
print('Sales workers have a relatively lower daily commuting distance ({}km) compared to other occupations.'.format(round(value, 2 )))
```
